# Ejercicio 1. Creación de una base de datos de películas extraídas de un API

### Fase 1: Extracción de datos de películas

```python
import requests
```

```python
import pandas as pd
```

```python
url_peliculas = f"https://beta.adalab.es/resources/apis/pelis/pelis.json"

llamada = requests.get(url_peliculas)

print(llamada.status_code)
```

```python
datos_peliculas = llamada.json()

datos_peliculas
```

```python
lista_peliculas = []

for pelicula in datos_peliculas:
    pelicula_info = {
        "titulo": pelicula["titulo"],
        "año de lanzamiento": pelicula["año"],
        "duración": pelicula["duracion"],
        "genero": pelicula["genero"],
        "adultos": pelicula["adultos"]
    }

    lista_peliculas.append(pelicula_info)
```

```python
lista_peliculas
```

```python
df_peliculas = pd.DataFrame(lista_peliculas)

df_peliculas                                 # quiero visualizar la informacion en formato tabla
```

```python
lista_peliculas_valores = []       # quiero los datos solos sin claves

for pelicula in lista_peliculas:
    info = [
        pelicula.get("titulo"),
        pelicula.get("año de lanzamiento"),
        pelicula.get("duración"),
        pelicula.get("genero"),
        pelicula.get("adultos")
    ]

    lista_peliculas_valores.append(info)

lista_peliculas_valores
```

```python
for fila in lista_peliculas_valores:
    print(tuple(fila))
```
